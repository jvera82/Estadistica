# Task
Estimar la altura promedio de estudiantes universitarios construyendo un intervalo de confianza del 95% para la media poblacional, dada una muestra de 50 estudiantes con una media de 170 cm y una desviación estándar poblacional de 10 cm. Adicionalmente, realizar una prueba de hipótesis completa siguiendo los pasos especificados, incluyendo la definición de hipótesis, nivel de significancia, identificación del estadístico de prueba, cálculo del valor crítico y región de rechazo, cálculo del estadístico de prueba y valor p, graficación de la distribución con las regiones relevantes, toma de decisión estadística y conclusión práctica con explicación del valor p. Utilizar `scipy.stats` para cálculos estadísticos y `matplotlib` para graficar, comentando todo el código y organizando las respuestas con texto interpretativo.

## Cargar librerías

### Subtask:
Importar las librerías necesarias (scipy.stats, matplotlib, numpy).


**Reasoning**:
Import the necessary libraries for statistical calculations and plotting.



```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

## Definir parámetros del problema

### Subtask:
Definir los valores dados en el enunciado (media muestral, tamaño de la muestra, desviación estándar poblacional, nivel de confianza).


**Reasoning**:
Assign the given values from the problem statement to variables.



```python
# Definir los valores dados en el enunciado
sample_mean = 170  # Media muestral en cm
sample_size = 50  # Tamaño de la muestra
population_std_dev = 10  # Desviación estándar poblacional en cm
confidence_level = 0.95  # Nivel de confianza
```

## Calcular el intervalo de confianza

### Subtask:
Calcular el intervalo de confianza para la media poblacional utilizando la fórmula adecuada para cuando la desviación estándar poblacional es conocida y obteniendo el valor crítico de Z con `scipy.stats`.


**Reasoning**:
Calculate the standard error, the critical Z-value, the margin of error, and the confidence interval bounds using the provided parameters and display the results.



```python
# 1. Calcular el error estándar de la media
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar: {standard_error:.4f}")

# 2. Calcular el valor crítico de Z para el nivel de confianza del 95%
# Para un intervalo de confianza del 95%, el área en cada cola es (1 - 0.95) / 2 = 0.025
# Necesitamos el valor Z tal que el área a la izquierda sea 1 - 0.025 = 0.975
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(f"Valor crítico de Z: {z_critical:.4f}")

# 3. Calcular el margen de error
margin_of_error = z_critical * standard_error
print(f"Margen de error: {margin_of_error:.4f}")

# 4. Calcular el límite inferior del intervalo de confianza
confidence_interval_lower = sample_mean - margin_of_error
print(f"Límite inferior del intervalo de confianza: {confidence_interval_lower:.4f}")

# 5. Calcular el límite superior del intervalo de confianza
confidence_interval_upper = sample_mean + margin_of_error
print(f"Límite superior del intervalo de confianza: {confidence_interval_upper:.4f}")

print(f"\nEl intervalo de confianza del {confidence_level*100}% para la media poblacional es: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")
```

## Definir hipótesis

### Subtask:
Establecer las hipótesis nula y alternativa para la prueba de hipótesis.


**Reasoning**:
Define the null and alternative hypotheses as string variables and then print them.



```python
# 1. Definir la hipótesis nula (H₀)
# En este caso, dado que estamos estimando la media, la hipótesis nula podría ser que la media poblacional
# es igual a la media muestral, o un valor relevante para el contexto del problema.
# Sin embargo, para una prueba de hipótesis estándar en esta situación, a menudo se plantea
# una hipótesis nula de que la media poblacional es igual a algún valor hipotetizado.
# Dado que no se proporciona un valor hipotetizado específico en el enunciado para la prueba,
# utilizaremos la media muestral como un punto de referencia para definir las hipótesis
# para ilustrar el proceso, aunque en una prueba real se usaría un valor predefinido.
# Para este ejercicio, asumiremos una hipótesis nula de que la media poblacional es 170 cm.
null_hypothesis = "H₀: La altura promedio de los estudiantes universitarios es igual a 170 cm (μ = 170)."

# 2. Definir la hipótesis alternativa (H₁)
# La hipótesis alternativa para una prueba de dos colas es que la media poblacional no es igual al valor de la hipótesis nula.
alternative_hypothesis = "H₁: La altura promedio de los estudiantes universitarios no es igual a 170 cm (μ ≠ 170)."

# 3. Imprimir las hipótesis
print("Hipótesis nula:", null_hypothesis)
print("Hipótesis alternativa:", alternative_hypothesis)
```

## Identificar estadístico de prueba y distribución

### Subtask:
Identificar el estadístico de prueba y su distribución bajo la hipótesis nula.


**Reasoning**:
Identify the appropriate test statistic and its distribution under the null hypothesis for a hypothesis test concerning the population mean when the population standard deviation is known.



```python
# Identificar el estadístico de prueba apropiado
# Dado que la desviación estándar poblacional (σ) es conocida y el tamaño de la muestra es suficientemente grande (n > 30),
# el estadístico de prueba apropiado para probar una hipótesis sobre la media poblacional (μ) es el estadístico Z.

# Determinar la distribución del estadístico de prueba bajo la hipótesis nula
# Bajo la hipótesis nula (H₀: μ = μ₀, donde μ₀ es el valor hipotetizado para la media poblacional),
# el estadístico Z sigue una distribución normal estándar (Z ~ N(0, 1)).

# Documentar el estadístico de prueba y su distribución
print("Estadístico de prueba: Z")
print("Distribución bajo la hipótesis nula: Distribución Normal Estándar (Z ~ N(0, 1))")

# Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)
# Donde:
# x̄ es la media muestral
# μ₀ es el valor de la media poblacional bajo la hipótesis nula
# σ es la desviación estándar poblacional
# n es el tamaño de la muestra
```

## Calcular el valor crítico y la región de rechazo

### Subtask:
Calcular el valor o los valores críticos de Z y definir la región o regiones de rechazo para la prueba de hipótesis, basándose en el nivel de significancia (α) y el tipo de prueba (dos colas).


**Reasoning**:
Calculate the critical Z-values for a two-tailed test and define the rejection region based on the significance level.



```python
# 1. Determinar el nivel de significancia para cada cola (prueba de dos colas)
alpha_per_tail = alpha / 2
print(f"Nivel de significancia por cola (α/2): {alpha_per_tail}")

# 2. Calcular el valor crítico de Z para el área de la cola inferior
z_critical_lower = stats.norm.ppf(alpha_per_tail)
print(f"Valor crítico de Z (cola inferior): {z_critical_lower}")

# 3. Determinar el valor crítico de Z para el área de la cola superior (debido a la simetría)
z_critical_upper = -z_critical_lower  # O stats.norm.ppf(1 - alpha_per_tail)
print(f"Valor crítico de Z (cola superior): {z_critical_upper}")

# 4. Definir verbalmente la región de rechazo
# Para una prueba de dos colas, la región de rechazo consiste en los valores del estadístico de prueba Z
# que son menores que el valor crítico inferior O mayores que el valor crítico superior.
rejection_region = f"La región de rechazo es Z < {z_critical_lower:.4f} o Z > {z_critical_upper:.4f}"

# 5. Imprimir los valores críticos de Z y la descripción de la región de rechazo
print("\nRegión de rechazo:")
print(rejection_region)
```

## Calcular el estadístico de prueba y el valor p

### Subtask:
Calcular el estadístico de prueba Z a partir de los datos de la muestra y el valor p asociado utilizando `scipy.stats`.


**Reasoning**:
Calculate the Z-test statistic and the associated p-value using the provided sample data and the standard normal distribution from scipy.stats, and then print these values.



```python
# 1. Calcular el estadístico de prueba Z
# La media hipotetizada (μ₀) bajo la hipótesis nula es 170 cm.
hypothesized_mean = 170
# Ya tenemos sample_mean = 170 y standard_error calculados previamente.
z_statistic = (sample_mean - hypothesized_mean) / standard_error

# 2. Calcular el valor p asociado al estadístico de prueba Z para una prueba de dos colas
# Para una distribución normal estándar, el valor p de dos colas es 2 * P(Z > |estadistico_z|)
# P(Z > |estadistico_z|) es 1 - P(Z <= |estadistico_z|), que es 1 - stats.norm.cdf(abs(estadistico_z))
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# 3. Imprimir el valor del estadístico de prueba Z calculado y el valor p asociado
print(f"Estadístico de prueba Z: {z_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")
```

## Graficar la distribución

### Subtask:
Graficar la distribución normal estándar, indicando la región de rechazo, el valor crítico y el estadístico de prueba calculado usando `matplotlib`.


**Reasoning**:
Generate x values, calculate the PDF, and plot the standard normal distribution with the critical values and the Z-statistic.



```python
# 1. Crear un rango de valores para el eje x
x = np.linspace(-4, 4, 1000)

# 2. Calcular la función de densidad de probabilidad (PDF) para estos valores x
pdf = stats.norm.pdf(x, 0, 1) # Media 0 y desviación estándar 1 para la distribución normal estándar

# 3. Crear un gráfico de la PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Normal Estándar')

# 4. Añadir líneas verticales en los valores críticos de Z
plt.axvline(z_critical_lower, color='red', linestyle='--', label=f'Valor Crítico Z (Inferior): {z_critical_lower:.4f}')
plt.axvline(z_critical_upper, color='red', linestyle='--', label=f'Valor Crítico Z (Superior): {z_critical_upper:.4f}')

# 5. Añadir una línea vertical en el valor del estadístico de prueba Z calculado
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z: {z_statistic:.4f}')

# 6. Sombrear las regiones de rechazo en el gráfico
# Sombrear la cola inferior
x_reject_lower = x[x < z_critical_lower]
pdf_reject_lower = stats.norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, pdf_reject_lower, color='red', alpha=0.3, label='Región de Rechazo')

# Sombrear la cola superior
x_reject_upper = x[x > z_critical_upper]
pdf_reject_upper = stats.norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, pdf_reject_upper, color='red', alpha=0.3) # No label para evitar duplicado

# 7. Añadir un título al gráfico y etiquetas a los ejes
plt.title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba Z')
plt.xlabel('Valores de Z')
plt.ylabel('Densidad de Probabilidad')

# 8. Añadir una leyenda
plt.legend()

# 9. Mostrar el gráfico
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 1. Comparar el valor p con el nivel de significancia (α)
print(f"Comparación del valor p ({p_value:.4f}) con el nivel de significancia (α = {alpha}):")
if p_value <= alpha:
    p_value_decision = "Rechazar la hipótesis nula."
else:
    p_value_decision = "No rechazar la hipótesis nula."
print(f"Decisión basada en el valor p: {p_value_decision}")

# 2. Comparar el estadístico de prueba Z con los valores críticos
print(f"\nComparación del estadístico de prueba Z ({z_statistic:.4f}) con los valores críticos Z ({z_critical_lower:.4f} y {z_critical_upper:.4f}):")
if z_statistic < z_critical_lower or z_statistic > z_critical_upper:
    z_statistic_decision = "Rechazar la hipótesis nula."
else:
    z_statistic_decision = "No rechazar la hipótesis nula."
print(f"Decisión basada en el estadístico de prueba Z: {z_statistic_decision}")

# 3. Tomar una decisión estadística basada en las comparaciones
# Ambas comparaciones deben llevar a la misma conclusión.
print("\nDecisión estadística final:")
if p_value_decision == z_statistic_decision:
    statistical_decision = p_value_decision
    print(statistical_decision)
else:
    statistical_decision = "Las decisiones basadas en el valor p y el estadístico Z no coinciden. Revisar cálculos."
    print(statistical_decision)


# 4. Redactar una conclusión práctica clara y concisa
print("\nConclusión práctica:")
if statistical_decision == "No rechazar la hipótesis nula.":
    practical_conclusion = (
        f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística "
        f"para rechazar la hipótesis nula de que la altura promedio de los estudiantes universitarios es igual a 170 cm. "
        f"El valor p ({p_value:.4f}) es mayor que el nivel de significancia ({alpha}), lo que indica "
        f"que si la altura promedio real fuera 170 cm, observar una media muestral de 170 cm "
        f"o más extrema (en cualquier dirección) sería muy probable (probabilidad = {p_value:.4f})."
    )
elif statistical_decision == "Rechazar la hipótesis nula.":
     practical_conclusion = (
        f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística "
        f"para rechazar la hipótesis nula de que la altura promedio de los estudiantes universitarios es igual a 170 cm. "
        f"El valor p ({p_value:.4f}) es menor o igual que el nivel de significancia ({alpha}), lo que indica "
        f"que si la altura promedio real fuera 170 cm, observar una media muestral de 170 cm "
        f"o más extrema (en cualquier dirección) sería muy improbable (probabilidad = {p_value:.4f})."
    )
else:
    practical_conclusion = statistical_decision # En caso de que las decisiones no coincidan

# 5. Imprimir la conclusión práctica
print(practical_conclusion)
```

## Calcular el estadístico de prueba y el valor p

### Subtask:
Calcular el estadístico de prueba Z a partir de los datos de la muestra y el valor p asociado utilizando `scipy.stats`.

**Reasoning**:
Calculate the Z-test statistic and the associated p-value using the provided sample data and the standard normal distribution from scipy.stats, and then print these values.

```python
# 1. Calculate the Z-test statistic
# The hypothesized mean (μ₀) under the null hypothesis is the value defined in the hypothesis step (hypothesized_mean).
# We use the previously calculated sample_mean and standard_error.
z_statistic = (sample_mean - hypothesized_mean) / standard_error

# 2. Calculate the two-tailed p-value associated with the Z-statistic
# The p-value for a two-tailed test is 2 * P(Z > |z_statistic|)
# This is equivalent to 2 * (1 - P(Z <= |z_statistic|)), which can be calculated using the cumulative distribution function (CDF).
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# 3. Print the calculated Z-statistic and the p-value
print(f"Estadístico de prueba Z: {z_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")
```

```python
# 1. Determinar el nivel de significancia para cada cola (prueba de dos colas)
# Dividir el nivel de significancia total (alpha) por 2.
alpha_per_tail = alpha / 2
print(f"Nivel de significancia por cola (α/2): {alpha_per_tail:.4f}")

# 2. Calcular el valor crítico de Z para el área de la cola inferior
# Usar stats.norm.ppf con el área de la cola inferior (alpha_per_tail).
z_critical_lower = stats.norm.ppf(alpha_per_tail)
print(f"Valor crítico de Z (cola inferior): {z_critical_lower:.4f}")

# 3. Determinar el valor crítico de Z para la cola superior
# Para una distribución normal estándar simétrica, el valor crítico superior es el negativo del valor crítico inferior.
z_critical_upper = -z_critical_lower
print(f"Valor crítico de Z (cola superior): {z_critical_upper:.4f}")

# 4. Definir verbalmente la región de rechazo
# La región de rechazo es donde el estadístico de prueba Z cae para rechazar H₀.
# Para una prueba de dos colas, es Z < z_critical_lower o Z > z_critical_upper.
rejection_region = f"La región de rechazo es Z < {z_critical_lower:.4f} o Z > {z_critical_upper:.4f}"

# 5. Imprimir la descripción de la región de rechazo
print("\nRegión de rechazo:")
print(rejection_region)
```

## Calcular el intervalo de confianza

### Subtask:
Calcular el intervalo de confianza del 95% para la duración media poblacional.

**Reasoning**:
Calculate the standard error, the critical Z-value, the margin of error, and the confidence interval bounds using the provided parameters and print the results.

```python
# 1. Calcular el error estándar de la media
# El error estándar es la desviación estándar poblacional dividida por la raíz cuadrada del tamaño de la muestra.
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar: {standard_error:.4f}")

# 2. Calcular el valor crítico de Z para un nivel de confianza del 95%
# Para un intervalo de confianza del 95%, el área en cada cola es (1 - 0.95) / 2 = 0.025
# Necesitamos el valor Z tal que el área a la izquierda sea 1 - 0.025 = 0.975
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(f"Valor crítico de Z: {z_critical:.4f}")

# 3. Calcular el margen de error
# El margen de error es el valor crítico de Z multiplicado por el error estándar.
margin_of_error = z_critical * standard_error
print(f"Margen de error: {margin_of_error:.4f}")

# 4. Calcular el límite inferior del intervalo de confianza
# El límite inferior es la media muestral menos el margen de error.
confidence_interval_lower = sample_mean - margin_of_error
print(f"Límite inferior del intervalo de confianza: {confidence_interval_lower:.4f}")

# 5. Calcular el límite superior del intervalo de confianza
# El límite superior es la media muestral más el margen de error.
confidence_interval_upper = sample_mean + margin_of_error
print(f"Límite superior del intervalo de confianza: {confidence_interval_upper:.4f}")

# 6. Imprimir el intervalo de confianza calculado de manera clara
print(f"\nEl intervalo de confianza del {confidence_level*100}% para la duración media poblacional es: ({confidence_interval_lower:.4f} horas, {confidence_interval_upper:.4f} horas)")
```

# Task
Determina un intervalo de confianza del 90 % para la media poblacional de las calificaciones de un examen nacional, dada una muestra aleatoria de 100 estudiantes con una calificación promedio de 75 puntos y una desviación estándar histórica de 15 puntos. Realiza una prueba de hipótesis completa, incluyendo la definición de hipótesis, nivel de significancia, identificación del estadístico de prueba, cálculo del valor crítico y región de rechazo, cálculo del estadístico de prueba y valor p, graficación de la distribución con las regiones relevantes, toma de decisión estadística y conclusión práctica. The subtask is to import the necessary libraries. This can be achieved by importing `scipy.stats`, `matplotlib.pyplot`, and `numpy`.



```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

## Definir parámetros del problema

### Subtask:
Definir los valores dados en el enunciado (media muestral, tamaño de la muestra, desviación estándar poblacional, nivel de confianza).

Assign the given values from the problem statement to variables and print them.



```python
# 1. Asignar el valor de la media muestral
sample_mean = 75  # Calificación promedio de la muestra

# 2. Asignar el valor del tamaño de la muestra
sample_size = 100  # Tamaño de la muestra

# 3. Asignar el valor de la desviación estándar poblacional
population_std_dev = 15  # Desviación estándar histórica de la población

# 4. Asignar el valor del nivel de confianza
confidence_level = 0.90  # Nivel de confianza del 90%

# 5. Imprimir los valores asignados
print(f"Media muestral: {sample_mean}")
print(f"Tamaño de la muestra: {sample_size}")
print(f"Desviación estándar poblacional: {population_std_dev}")
print(f"Nivel de confianza: {confidence_level}")
```

## Calcular el intervalo de confianza

### Subtask:
Calcular el intervalo de confianza del 90% para la media poblacional.


**Reasoning**:
Calculate the standard error, the critical Z-value, the margin of error, and the confidence interval bounds using the provided parameters and print the results.



```python
# 1. Calculate the standard error of the mean
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar: {standard_error:.4f}")

# 2. Calculate the critical Z-value for a 90% confidence level
# For a 90% confidence interval, the area in each tail is (1 - 0.90) / 2 = 0.05
# We need the Z-value such that the area to the left is 1 - 0.05 = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(f"Valor crítico de Z: {z_critical:.4f}")

# 3. Calculate the margin of error
margin_of_error = z_critical * standard_error
print(f"Margen de error: {margin_of_error:.4f}")

# 4. Calculate the lower bound of the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
print(f"Límite inferior del intervalo de confianza: {confidence_interval_lower:.4f}")

# 5. Calculate the upper bound of the confidence interval
confidence_interval_upper = sample_mean + margin_of_error
print(f"Límite superior del intervalo de confianza: {confidence_interval_upper:.4f}")

# 6. Print the confidence interval
print(f"\nEl intervalo de confianza del {confidence_level*100}% para la media poblacional es: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")
```

```python
# 1. Define the significance level (alpha)
# Based on common practice and the 90% confidence level used for the interval,
# a typical choice for a two-tailed test is 1 - confidence_level.
alpha = 1 - confidence_level

# 2. Print the defined significance level
print(f"Nivel de significancia (α): {alpha}")
```

## Identificar estadístico de prueba y distribución

### Subtask:
Identificar el estadístico de prueba y su distribución bajo la hipótesis nula.


**Reasoning**:
Based on the problem description (sample size > 30 and population standard deviation known), the appropriate test statistic for a hypothesis test about the population mean is the Z-statistic, which follows a standard normal distribution under the null hypothesis.



```python
# 1. Identify the appropriate test statistic
# Given that the population standard deviation (σ) is known and the sample size (n = 100) is large,
# the appropriate test statistic for testing a hypothesis about the population mean (μ) is the Z-statistic.

# 2. Determine the distribution of the test statistic under the null hypothesis
# Under the null hypothesis (H₀), the Z-statistic follows a standard normal distribution with a mean of 0 and a standard deviation of 1 (Z ~ N(0, 1)).

# 3. Print the identified test statistic and its distribution
print("Estadístico de prueba: Z")
print("Distribución bajo la hipótesis nula: Distribución Normal Estándar (Z ~ N(0, 1))")

# Formula for the Z-statistic: Z = (x̄ - μ₀) / (σ / √n)
# Where:
# x̄ is the sample mean
# μ₀ is the hypothesized population mean under the null hypothesis
# σ is the population standard deviation
# n is the sample size
```

## Calcular el valor crítico y la región de rechazo

### Subtask:
Calcular el valor o los valores críticos de Z y definir la región o regiones de rechazo para la prueba de hipótesis, basándose en el nivel de significancia (α) y el tipo de prueba (dos colas).


**Reasoning**:
Calculate the critical Z-values for a two-tailed test and define the rejection region based on the significance level.



```python
# 1. Determinar el nivel de significancia para cada cola (prueba de dos colas)
# Dividir el nivel de significancia total (alpha) por 2.
alpha_per_tail = alpha / 2
print(f"Nivel de significancia por cola (α/2): {alpha_per_tail}")

# 2. Calcular el valor crítico de Z para el área de la cola inferior
# Usar stats.norm.ppf con el área de la cola inferior (alpha_per_tail).
z_critical_lower = stats.norm.ppf(alpha_per_tail)
print(f"Valor crítico de Z (cola inferior): {z_critical_lower:.4f}")

# 3. Determinar el valor crítico de Z para la cola superior
# Para una distribución normal estándar simétrica, el valor crítico superior es el negativo del valor crítico inferior.
z_critical_upper = -z_critical_lower
print(f"Valor crítico de Z (cola superior): {z_critical_upper:.4f}")

# 4. Definir verbalmente la región de rechazo
# La región de rechazo es donde el estadístico de prueba Z cae para rechazar H₀.
# Para una prueba de dos colas, es Z < z_critical_lower o Z > z_critical_upper.
rejection_region = f"La región de rechazo es Z < {z_critical_lower:.4f} o Z > {z_critical_upper:.4f}"

# 5. Imprimir la descripción de la región de rechazo
print("\nRegión de rechazo:")
print(rejection_region)
```

## Calcular el estadístico de prueba y el valor p

### Subtask:
Calcular el estadístico de prueba Z a partir de los datos de la muestra y el valor p asociado utilizando `scipy.stats`.


**Reasoning**:
Calculate the Z-test statistic and the associated p-value using the provided sample data and the standard normal distribution from scipy.stats, and then print these values.



```python
# 1. Calculate the Z-test statistic
# The hypothesized mean (μ₀) under the null hypothesis is 75 points, as defined in the hypothesis step.
# We already have sample_mean = 75 and standard_error = 1.5 calculated previously.
z_statistic = (sample_mean - hypothesized_mean_for_test) / standard_error

# 2. Calculate the two-tailed p-value associated with the Z-statistic
# The p-value for a two-tailed test is 2 * P(Z > |z_statistic|)
# P(Z > |z_statistic|) is 1 - P(Z <= |z_statistic|), which is 1 - stats.norm.cdf(abs(z_statistic))
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# 3. Print the calculated Z-statistic and the p-value
print(f"Estadístico de prueba Z: {z_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")
```

```python
# 1. Create a range of x-values for plotting the standard normal distribution
x = np.linspace(-4, 4, 1000)

# 2. Calculate the probability density function (PDF) for these x-values
# Use the standard normal distribution (mean=0, standard deviation=1) from scipy.stats.norm.pdf
pdf = stats.norm.pdf(x, 0, 1)

# 3. Create a plot using matplotlib.pyplot and plot the standard normal distribution PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Normal Estándar')

# 4. Add vertical lines on the plot at the calculated critical Z-values
# Style them appropriately (e.g., dashed lines, red color) and add labels.
plt.axvline(z_critical_lower, color='red', linestyle='--', label=f'Valor Crítico Z (Inferior): {z_critical_lower:.4f}')
plt.axvline(z_critical_upper, color='red', linestyle='--', label=f'Valor Crítico Z (Superior): {z_critical_upper:.4f}')

# 5. Add a vertical line on the plot at the calculated Z-test statistic
# Style it differently (e.g., solid line, green color) and add a label.
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z: {z_statistic:.4f}')

# 6. Shade the rejection regions on the plot.
# These are the areas under the PDF curve to the left of z_critical_lower and to the right of z_critical_upper.
# Use the fill_between function and an appropriate color and transparency. Add a label for the rejection region.
x_reject_lower = x[x < z_critical_lower]
pdf_reject_lower = stats.norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, pdf_reject_lower, color='red', alpha=0.3, label='Región de Rechazo')

x_reject_upper = x[x > z_critical_upper]
pdf_reject_upper = stats.norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, pdf_reject_upper, color='red', alpha=0.3) # No label needed for the second part

# 7. Add a title to the plot and labels to the x and y axes.
plt.title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba Z')
plt.xlabel('Valores de Z')
plt.ylabel('Densidad de Probabilidad')

# 8. Add a legend to the plot to identify the different elements.
plt.legend()

# 9. Add a grid to the plot for better readability.
plt.grid(True, linestyle='--', alpha=0.6)

# 10. Display the plot.
plt.show()
```

```python
# 1. Calcular el error estándar de la media
# El error estándar es la desviación estándar poblacional dividida por la raíz cuadrada del tamaño de la muestra.
standard_error = population_std_dev / np.sqrt(sample_size)
print(f"Error estándar: {standard_error:.4f}")

# 2. Calcular el valor crítico de Z para el nivel de confianza del 99%
# Para un intervalo de confianza del 99%, el área en cada cola es (1 - 0.99) / 2 = 0.005
# Necesitamos el valor Z tal que el área a la izquierda sea 1 - 0.005 = 0.995
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(f"Valor crítico de Z: {z_critical:.4f}")

# 3. Calcular el margen de error
# El margen de error es el valor crítico de Z multiplicado por el error estándar.
margin_of_error = z_critical * standard_error
print(f"Margen de error: {margin_of_error:.4f}")

# 4. Calcular el límite inferior del intervalo de confianza
# El límite inferior es la media muestral menos el margen de error.
confidence_interval_lower = sample_mean - margin_of_error
print(f"Límite inferior del intervalo de confianza: {confidence_interval_lower:.4f}")

# 5. Calcular el límite superior del intervalo de confianza
# El límite superior es la media muestral más el margen de error.
confidence_interval_upper = sample_mean + margin_of_error
print(f"Límite superior del intervalo de confianza: {confidence_interval_upper:.4f}")

# 6. Imprimir el intervalo de confianza calculado de manera clara
print(f"\nEl intervalo de confianza del {confidence_level*100}% para el diámetro promedio real es: ({confidence_interval_lower:.4f} cm, {confidence_interval_upper:.4f} cm)")
```

**Reasoning**:
Identify the appropriate test statistic and its distribution under the null hypothesis for a hypothesis test concerning the population mean when the population standard deviation is known.



```python
# 1. Identify the appropriate test statistic
# Given that the population standard deviation (σ) is known and the sample size (n = 30) is sufficiently large (although technically n=30 is the threshold, with known sigma Z-test is appropriate),
# the appropriate test statistic for testing a hypothesis about the population mean (μ) is the Z-statistic.

# 2. Determine the distribution of the test statistic under the null hypothesis
# Under the null hypothesis (H₀), the Z-statistic follows a standard normal distribution with a mean of 0 and a standard deviation of 1 (Z ~ N(0, 1)).

# 3. Print the identified test statistic and its distribution
print("Estadístico de prueba: Z")
print("Distribución bajo la hipótesis nula: Distribución Normal Estándar (Z ~ N(0, 1))")

# Formula for the Z-statistic: Z = (x̄ - μ₀) / (σ / √n)
# Where:
# x̄ is the sample mean
# μ₀ is the hypothesized population mean under the null hypothesis
# σ is the population standard deviation
# n is the sample size
```

**Reasoning**:
Calculate the critical Z-values for a two-tailed test and define the rejection region based on the significance level.



```python
# 1. Determinar el nivel de significancia para cada cola (prueba de dos colas)
# Dividir el nivel de significancia total (alpha) por 2.
alpha_per_tail = alpha / 2
print(f"Nivel de significancia por cola (α/2): {alpha_per_tail:.4f}")

# 2. Calcular el valor crítico de Z para el área de la cola inferior
# Usar stats.norm.ppf con el área de la cola inferior (alpha_per_tail).
z_critical_lower = stats.norm.ppf(alpha_per_tail)
print(f"Valor crítico de Z (cola inferior): {z_critical_lower:.4f}")

# 3. Determinar el valor crítico de Z para la cola superior
# Para una distribución normal estándar simétrica, el valor crítico superior es el negativo del valor crítico inferior.
z_critical_upper = -z_critical_lower
print(f"Valor crítico de Z (cola superior): {z_critical_upper:.4f}")

# 4. Definir verbalmente la región de rechazo
# La región de rechazo es donde el estadístico de prueba Z cae para rechazar H₀.
# Para una prueba de dos colas, es Z < z_critical_lower o Z > z_critical_upper.
rejection_region = f"La región de rechazo es Z < {z_critical_lower:.4f} o Z > {z_critical_upper:.4f}"

# 5. Imprimir la descripción de la región de rechazo
print("\nRegión de rechazo:")
print(rejection_region)
```

**Reasoning**:
Calculate the Z-test statistic and the associated p-value using the provided sample data and the standard normal distribution from scipy.stats, and then print these values.



```python
# 1. Calculate the Z-test statistic
# The hypothesized mean (μ₀) under the null hypothesis is the value defined in the hypothesis step (hypothesized_mean).
# We use the previously calculated sample_mean and standard_error.
z_statistic = (sample_mean - hypothesized_mean) / standard_error

# 2. Calculate the two-tailed p-value associated with the Z-statistic
# The p-value for a two-tailed test is 2 * P(Z > |z_statistic|)
# This is equivalent to 2 * (1 - P(Z <= |z_statistic|)), which can be calculated using the cumulative distribution function (CDF).
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# 3. Print the calculated Z-statistic and the p-value
print(f"Estadístico de prueba Z: {z_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")
```

```python
# 1. Create a range of x-values for plotting the standard normal distribution
# Use a range that covers the relevant part of the distribution, for instance from -4 to 4 standard deviations.
x = np.linspace(-4, 4, 1000)

# 2. Calculate the probability density function (PDF) for these x-values
# Use the standard normal distribution (mean=0, standard deviation=1) from scipy.stats.norm.pdf.
pdf = stats.norm.pdf(x, 0, 1)

# 3. Create a plot using matplotlib.pyplot and plot the standard normal distribution PDF
# Set the figure size for better readability.
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Normal Estándar')

# 4. Add vertical lines on the plot at the calculated critical Z-values
# Style them appropriately (e.g., dashed lines, red color) and add labels indicating their values.
plt.axvline(z_critical_lower, color='red', linestyle='--', label=f'Valor Crítico Z (Inferior): {z_critical_lower:.4f}')
plt.axvline(z_critical_upper, color='red', linestyle='--', label=f'Valor Crítico Z (Superior): {z_critical_upper:.4f}')

# 5. Add a vertical line on the plot at the calculated Z-test statistic
# Style it differently (e.g., solid line, green color) and add a label indicating its value.
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico de Prueba Z: {z_statistic:.4f}')

# 6. Shade the rejection regions on the plot.
# These are the areas under the PDF curve to the left of z_critical_lower and to the right of z_critical_upper.
# Use the fill_between function and an appropriate color and transparency (alpha). Add a label for the rejection region.
x_reject_lower = x[x < z_critical_lower]
pdf_reject_lower = stats.norm.pdf(x_reject_lower, 0, 1)
plt.fill_between(x_reject_lower, 0, pdf_reject_lower, color='red', alpha=0.3, label='Región de Rechazo')

x_reject_upper = x[x > z_critical_upper]
pdf_reject_upper = stats.norm.pdf(x_reject_upper, 0, 1)
plt.fill_between(x_reject_upper, 0, pdf_reject_upper, color='red', alpha=0.3) # No label needed for the second part

# 7. Add a title to the plot and labels to the x and y axes for clarity.
plt.title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba Z')
plt.xlabel('Valores de Z')
plt.ylabel('Densidad de Probabilidad')

# 8. Add a legend to the plot to identify the different elements.
plt.legend()

# 9. Add a grid to the plot for better readability. Use dashed lines and some transparency.
plt.grid(True, linestyle='--', alpha=0.6)

# 10. Display the plot.
plt.show()
```

**Reasoning**:
Identify the appropriate test statistic and its distribution under the null hypothesis for a hypothesis test concerning the population mean when the population standard deviation is known.



```python
# 1. Identify the appropriate test statistic
# Given that the population standard deviation (σ) is known and the sample size (n = 40) is sufficiently large (n >= 30),
# the appropriate test statistic for testing a hypothesis about the population mean (μ) is the Z-statistic.

# 2. Determine the distribution of the test statistic under the null hypothesis
# Under the null hypothesis (H₀), the Z-statistic follows a standard normal distribution with a mean of 0 and a standard deviation of 1 (Z ~ N(0, 1)).

# 3. Print the identified test statistic and its distribution
print("Estadístico de prueba: Z")
print("Distribución bajo la hipótesis nula: Distribución Normal Estándar (Z ~ N(0, 1))")

# Formula for the Z-statistic: Z = (x̄ - μ₀) / (σ / √n)
# Where:
# x̄ is the sample mean
# μ₀ is the hypothesized population mean under the null hypothesis
# σ is the population standard deviation
# n is the sample size
```

**Reasoning**:
Calculate the critical Z-values for a two-tailed test and define the rejection region based on the significance level.



```python
# 1. Determinar el nivel de significancia para cada cola (prueba de dos colas)
# Dividir el nivel de significancia total (alpha) por 2.
alpha_per_tail = alpha / 2
print(f"Nivel de significancia por cola (α/2): {alpha_per_tail:.4f}")

# 2. Calcular el valor crítico de Z para el área de la cola inferior
# Usar stats.norm.ppf con el área de la cola inferior (alpha_per_tail).
z_critical_lower = stats.norm.ppf(alpha_per_tail)
print(f"Valor crítico de Z (cola inferior): {z_critical_lower:.4f}")

# 3. Determinar el valor crítico de Z para la cola superior
# Para una distribución normal estándar simétrica, el valor crítico superior es el negativo del valor crítico inferior.
z_critical_upper = -z_critical_lower
print(f"Valor crítico de Z (cola superior): {z_critical_upper:.4f}")

# 4. Definir verbalmente la región de rechazo
# La región de rechazo es donde el estadístico de prueba Z cae para rechazar H₀.
# Para una prueba de dos colas, es Z < z_critical_lower o Z > z_critical_upper.
rejection_region = f"La región de rechazo es Z < {z_critical_lower:.4f} o Z > {z_critical_upper:.4f}"

# 5. Imprimir la descripción de la región de rechazo
print("\nRegión de rechazo:")
print(rejection_region)
```

**Reasoning**:
Calculate the Z-test statistic and the associated p-value using the provided sample data and the standard normal distribution from scipy.stats, and then print these values.



```python
# 1. Calculate the Z-test statistic
# The hypothesized mean (μ₀) under the null hypothesis is the value defined in the hypothesis step (hypothesized_mean).
# We use the previously calculated sample_mean and standard_error.
z_statistic = (sample_mean - hypothesized_mean) / standard_error

# 2. Calculate the two-tailed p-value associated with the Z-statistic
# The p-value for a two-tailed test is 2 * P(Z > |z_statistic|)
# This is equivalent to 2 * (1 - P(Z <= |z_statistic|)), which can be calculated using the cumulative distribution function (CDF).
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# 3. Print the calculated Z-statistic and the p-value
print(f"Estadístico de prueba Z: {z_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")
```

```python
# 1. Calcular los grados de libertad
degrees_of_freedom = sample_size - 1
print(f"Grados de libertad: {degrees_of_freedom}")

# 2. Calcular el error estándar de la media muestral
# Usar la desviación estándar muestral (sample_std_dev) ya que no conocemos la poblacional.
standard_error = sample_std_dev / np.sqrt(sample_size)
print(f"Error estándar: {standard_error:.4f}°C")

# 3. Calcular el valor crítico de t para el nivel de confianza del 95% y los grados de libertad
# Para un intervalo de confianza del 95% de dos colas, el área en cada cola es (1 - 0.95) / 2 = 0.025.
# Usar stats.t.ppf para encontrar el valor t. Necesitamos el valor t tal que el área a la izquierda sea 1 - 0.025 = 0.975.
t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)
print(f"Valor crítico de t: {t_critical:.4f}")

# 4. Calcular el margen de error
# El margen de error es el valor crítico de t multiplicado por el error estándar.
margin_of_error = t_critical * standard_error
print(f"Margen de error: {margin_of_error:.4f}°C")

# 5. Calcular el límite inferior del intervalo de confianza
# El límite inferior es la media muestral menos el margen de error.
confidence_interval_lower = sample_mean - margin_of_error
print(f"Límite inferior del intervalo de confianza: {confidence_interval_lower:.4f}°C")

# 6. Calcular el límite superior del intervalo de confianza
# El límite superior es la media muestral más el margen de error.
confidence_interval_upper = sample_mean + margin_of_error
print(f"Límite superior del intervalo de confianza: {confidence_interval_upper:.4f}°C")

# 7. Imprimir el intervalo de confianza calculado de manera clara
print(f"\nEl intervalo de confianza del {confidence_level*100}% para la temperatura media poblacional del café es: ({confidence_interval_lower:.4f}°C, {confidence_interval_upper:.4f}°C)")
```

## Calcular el intervalo de confianza (distribución t)

### Subtask:
Calcular el intervalo de confianza del 95% para el peso medio poblacional utilizando la distribución t.

**Reasoning**:
Calculate the degrees of freedom, standard error, critical t-value, margin of error, and the confidence interval bounds using the provided parameters and display the results.

```python
# 1. Calcular los grados de libertad
degrees_of_freedom = sample_size - 1
print(f"Grados de libertad: {degrees_of_freedom}")

# 2. Calcular el error estándar de la media muestral
# Usar la desviación estándar muestral (sample_std_dev) ya que no conocemos la poblacional.
standard_error = sample_std_dev / np.sqrt(sample_size)
print(f"Error estándar: {standard_error:.4f} kg")

# 3. Calcular el valor crítico de t para el nivel de confianza del 95% y los grados de libertad
# Para un intervalo de confianza del 95% de dos colas, el área en cada cola es (1 - 0.95) / 2 = 0.025.
# Necesitamos el valor t tal que el área a la izquierda sea 1 - 0.025 = 0.975.
t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)
print(f"Valor crítico de t: {t_critical:.4f}")

# 4. Calcular el margen de error
# El margen de error es el valor crítico de t multiplicado por el error estándar.
margin_of_error = t_critical * standard_error
print(f"Margen de error: {margin_of_error:.4f} kg")

# 5. Calcular el límite inferior del intervalo de confianza
# El límite inferior es la media muestral menos el margen de error.
confidence_interval_lower = sample_mean - margin_of_error
print(f"Límite inferior del intervalo de confianza: {confidence_interval_lower:.4f} kg")

# 6. Calcular el límite superior del intervalo de confianza
# El límite superior es la media muestral más el margen de error.
confidence_interval_upper = sample_mean + margin_of_error
print(f"Límite superior del intervalo de confianza: {confidence_interval_upper:.4f} kg")

# 7. Imprimir el intervalo de confianza calculado de manera clara
print(f"\nEl intervalo de confianza del {confidence_level*100}% para el peso medio poblacional es: ({confidence_interval_lower:.4f} kg, {confidence_interval_upper:.4f} kg)")
```

## Calcular el tamaño de muestra para estimar una media (σ conocida)

### Subtask:
Definir los parámetros del problema (desviación estándar poblacional, margen de error, nivel de confianza) y calcular el tamaño de muestra necesario.

**Reasoning**:
Define the given parameters and calculate the required sample size using the formula for estimating a population mean when the population standard deviation is known.

```python
import scipy.stats as stats
import numpy as np

# 1. Definir la desviación estándar poblacional (σ)
population_std_dev = 12  # Desviación estándar poblacional en puntos

# 2. Definir el margen de error deseado (E)
margin_of_error = 3  # Margen de error en puntos

# 3. Definir el nivel de confianza
confidence_level = 0.90  # Nivel de confianza del 90%

# 4. Calcular el valor crítico de Z para el nivel de confianza del 90%
# Para un nivel de confianza del 90%, el área en cada cola es (1 - 0.90) / 2 = 0.05.
# Necesitamos el valor Z tal que el área a la izquierda sea 1 - 0.05 = 0.95.
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(f"Valor crítico de Z: {z_critical:.4f}")

# 5. Calcular el tamaño de muestra necesario (n) utilizando la fórmula:
# n = (Z² * σ²) / E²
# Donde:
# Z es el valor crítico de Z
# σ es la desviación estándar poblacional
# E es el margen de error
n = (z_critical**2 * population_std_dev**2) / (margin_of_error**2)

# 6. Redondear hacia arriba al entero más cercano, ya que el tamaño de la muestra debe ser un número entero.
sample_size = np.ceil(n)

# 7. Imprimir el tamaño de muestra calculado
print(f"\nTamaño de muestra necesario: {int(sample_size)}")
```

## Calcular el tamaño de muestra para estimar una proporción (con y sin prueba preliminar)

### Subtask:
Definir los parámetros del problema y calcular el tamaño de muestra necesario con y sin la información de la prueba preliminar.

**Reasoning**:
Define the given parameters, calculate the required sample size using the formula for estimating a population proportion with and without a prior estimate, and summarize the results.

```python
import scipy.stats as stats
import numpy as np

# --- Parámetros del problema ---
# Proporción estimada de la prueba preliminar (p̂)
p_hat = 0.69  # 69% como decimal

# Margen de error deseado (E)
margin_of_error = 0.04  # 4% como decimal

# Nivel de confianza
confidence_level = 0.95  # 95% como decimal

print("--- Parámetros Definidos ---")
print(f"Proporción estimada de prueba preliminar (p̂): {p_hat}")
print(f"Margen de error (E): {margin_of_error}")
print(f"Nivel de confianza: {confidence_level}")

# --- Cálculo del tamaño de muestra con prueba preliminar ---
# Calcular el valor crítico de Z para el nivel de confianza del 95%
# Para un nivel de confianza del 95%, el área en cada cola es (1 - 0.95) / 2 = 0.025.
# Necesitamos el valor Z tal que el área a la izquierda sea 1 - 0.025 = 0.975.
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(f"\nValor crítico de Z para {confidence_level*100}% de confianza: {z_critical:.4f}")

# Calcular el tamaño de muestra necesario (n) utilizando la fórmula:
# n = (Z² * p̂ * (1 - p̂)) / E²
n_preliminar = (z_critical**2 * p_hat * (1 - p_hat)) / (margin_of_error**2)

# Redondear hacia arriba al entero más cercano
sample_size_preliminar = np.ceil(n_preliminar)

print("\n--- Tamaño de Muestra (con prueba preliminar) ---")
print(f"Tamaño de muestra necesario (con p̂ = {p_hat}): {int(sample_size_preliminar)}")

# --- Cálculo del tamaño de muestra siendo conservador (sin prueba preliminar) ---
# Asumir la proporción más conservadora (p = 0.5)
p_conservative = 0.5

# Calcular el tamaño de muestra necesario (n) utilizando la fórmula:
# n = (Z² * p * (1 - p)) / E²
n_conservative = (z_critical**2 * p_conservative * (1 - p_conservative)) / (margin_of_error**2)

# Redondear hacia arriba al entero más cercano
sample_size_conservative = np.ceil(n_conservative)

print("\n--- Tamaño de Muestra (siendo conservador) ---")
print(f"Tamaño de muestra necesario (con p = {p_conservative}): {int(sample_size_conservative)}")

# --- Resumen y Comparación ---
print("\n--- Resumen y Comparación ---")
print(f"Para estimar la proporción de pacientes que mejoran con el tratamiento con un margen de error del {margin_of_error*100}% y un nivel de confianza del {confidence_level*100}%:")
print(f"- Si se utiliza la proporción de la prueba preliminar ({p_hat*100}%), el tamaño de muestra necesario es aproximadamente {int(sample_size_preliminar)}.")
print(f"- Si se desea ser conservador (asumiendo 50%), el tamaño de muestra necesario es aproximadamente {int(sample_size_conservative)}.")
print("\nLa diferencia muestra cómo una estimación previa puede reducir el tamaño de muestra necesario.")
```
